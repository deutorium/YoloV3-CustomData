```python
from PIL import Image
import os, sys

path = "/home/sahil/pikachu/"
dirs = os.listdir( path )

def resize_images():
    for item in dirs:
        if os.path.isfile(path+item):
            im = Image.open(path+item)
            f, e = os.path.splitext(path+item)
            imResize = im.resize((64,64), Image.ANTIALIAS)
            imResize.save(path+'resized/'+item, 'JPEG', quality=90)

resize_images()
```

Rename file names(Preferably in sorted way)

```python
import os
import glob
  
def rename(): 
    path = "/home/sahil/pikachu/resized/"
    for count, filename in enumerate(sorted(os.listdir(path))):
        dst = str(count+1) + ".jpg"
        src =path + filename 
        dst =path + dst 
        os.rename(src, dst) 

rename()
```

Create custom.shapes with image dimensions 
We don't need this as we have already resized the images to 64 x 64

```python
from PIL import Image
import os, sys
path = "/home/sahil/pikachu/resized/"
file = open("/home/sahil/pikachu/custom.txt")
custom_shapes = open("/home/sahil/pikachu/custom.shapes","w")

for line in file:
    im = Image.open(path+line.split('/')[-1].rstrip())
    width, height = im.size
    custom_shapes.write(str(width)+" "+str(height)+"\n")
custom_shapes.close()
```

Run this code anyway. It will create a custom.shapes file containing the size of each pic in same order as custom.txt
